### Identifica problemas de calidad de datos (registros duplicados, valores faltantes, outliers, etc).

```python
#Inconsistent data
#The first think that we can see is that there are not null values at the begining, 
#but the reason for that is because some values represent the nulls with a ?
nulls = df.isnull().any()
columnas_con_nulos = nulls[nulls].index.tolist()
print("Columnas con valores nulos:", columnas_con_nulos)
#Firstly we change that character to a numpy null.
df.replace("?", np.nan, inplace=True)
print('After transformation:\n')
nulls = df.isnull().any()
columnas_con_nulos = nulls[nulls].index.tolist()
print("Columnas con valores nulos:", columnas_con_nulos)
```

```python
#Nulls
#Here we can see those columns that have null values, and the number 
#of them compare to the total amount of vals.
nulls = df.columns[df.isnull().any()]
columna_nulls=[]
for columna in nulls:
    nulos = df[columna].isnull().sum()
    total_datos = len(df.index)
    print(f"Columna: {columna}")
    columna_nulls.append(columna)
    print(f"Cantidad de nulos: {nulos}")
    print(f"Total de datos: {total_datos}")
    print("-" * 20)
```

```python
#The next thing to check is if in our dataset are duplicates
duplicates = df[df.duplicated()]

print("Registros duplicados:")
print(duplicates)
```

```python
#Outliers using box plots
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Calcula los límites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecta outliers
outliers = ((df < lower_bound) | (df > upper_bound)).sum()

# Imprime la cantidad de outliers por columna
print("Cantidad de outliers por columna:")
print(outliers)
```

#### Variables categoricas
Nominales:
En el caso de las variables categoricas, aquellas variables que tengan un numero menor a 5 categorías diferentes utilizaría one hot encoding para poder analizar los datos segun si presentan o no alguna caracteristica. 
En el caso especifico del numero de puertas lo mas conveniente sería hacer un 'cast' para cambiar el string a un integer y así poder aprovechar de mejor manera esta columna.
En el caso de la marca del vehículo dejaría la columna tal cual como esta, ya que me sirve para poder filtrar los datos y poder analizarlos respecto a cierta marca o si se ocupa realizar alguna tecnica de clasificación las convertiria a factor para así poder llevar a cabo la tecnica.
Ordinales:
En este dataset la unica variable categorica ordinal ya se encuentra con un tipo de dato de integer, el cual se entiende perfectamente, por lo que no es necesario llevar a cabo ningun cambio.

```python
#The categorical data groups the information into distinct fields
#df = pd.read_csv("G:/Mi unidad/IA/Estadistica/data_sets/Automobile.csv")
categorical = df[['make','fuel-type','aspiration','body-style','drive-wheels','engine-location','engine-type',	'fuel-system']]

categorical_encoded = pd.get_dummies(categorical)


# Ahora categorical_encoded contiene las columnas categóricas codificadas en formato one-hot
print(categorical_encoded.head())

df = df.drop(['make','fuel-type','aspiration','body-style','drive-wheels','engine-location','engine-type',	'fuel-system'], axis=1)

df = pd.concat([df, categorical_encoded], axis=1)

df.head(10)
```

#### Variables numericas:
En el caso de las variables numericas lo correcto sería modificar el tipo de dato a integer o float segun corresponda, ya que en el dataset algunas variables numericas aparecen como objetos.

```python
print('Before transformation: \n' ,df.dtypes)
```

```python
df['normalized-losses'] = df['normalized-losses'].astype('Int64')
df['horsepower']=df['horsepower'].astype('Int64')
df['bore'] = pd.to_numeric(df['bore'], errors='coerce')
df['bore']=df['bore'].astype('Float64')
df['price']=df['price'].astype('Int64')
df['peak-rpm']=df['peak-rpm'].astype('Int64')
df['stroke'] = pd.to_numeric(df['stroke'], errors='coerce')
df['stroke']=df['stroke'].astype('Float64')

print('After transformation: \n', df.dtypes)
```

#### Wrong Format   
En el caso de las dos variables (numero de puertas y numero de cilindros), es necesario realizar una transformación para reemplazar el valor en string por su valor numerico lo que facilitaria el analisis.

```python
mapping = {
    'one':1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}
df['number-of-doors'] = df['number-of-doors'].replace(mapping)
df['num-of-cylinders'] = df['num-of-cylinders'].replace(mapping)
df['number-of-doors'] = df['number-of-doors'].astype('Int64')
df['num-of-cylinders'] = df['num-of-cylinders'].astype('Int64')
df['symboling']=df['symboling'].astype('Int64')
df[['number-of-doors','num-of-cylinders']].head(10)
```

#### Nulos
En el caso de los nulos existen columnas con un numero de nulos menor a 10, y una sola columna con nulos mayor a 10; en ambos casos las columnas son numericas o se pueden convertir a numericas. 
Para aquellas columnas con un numero de nulos menor a 10 eliminaría los registros de aquellas variables utiles para mi analisis. Puesto que no afectaría perder menos del 10% de los registros.
Para la columna que contiene un gran numero de valores faltantes, si la columna no es de utilidad la eliminaría, si es de utilidad, rellenaría los valores faltantes de los nulos con la media según sea la marca del vehículo, con el objetivo de aproximarme lo mejor posible al valor del campo.

```python
media = df['normalized-losses'].mean()

# Rellenar los valores nulos con la media
df['normalized-losses'].fillna(media, inplace=True)

df_filtrado = df.dropna()

nulls = df_filtrado.columns[df_filtrado.isnull().any()]
columna_nulls=[]
for columna in nulls:
    nulos = df_filtrado[columna].isnull().sum()
    total_datos = len(df_filtrado.index)
    print(f"Columna: {columna}")
    columna_nulls.append(columna)
    print(f"Cantidad de nulos: {nulos}")
    print(f"Total de datos: {total_datos}")
    print("-" * 20)
print('Cantidad de datos originales: ',len(df.index),'\nCantidad de datos restantes: ',len(df_filtrado.index))
```

#### Outliers
En el caso de los outliers seguiría una metodología similar que con los nulos, eliminando aquellos registros de las columnas que presenten un numero menor a 10 outliers.
Para las columnas que sobrepasen dicho rango, buscaría emplear una técnica de transformación logaritmica o box-cox para reducir la influencia de los outliers, y en el dado caso que queden pocos eliminar los registros, si esta técnica no funciona cambiaría los valores outliers por valores limites según el rango intercuartilico que sobrepasen, de esta forma evito la eliminación de un gran numero de valores que pudiera llegar a afectar el analisis.
